##### 1. Определить (с обоснованием), зависимы или независимы следующие события:
- a. Несовместные события
- b. События, образующие σ-алгебру Σ в пространстве (Ω,Σ,P)
- c. События, имеющие одинаковую вероятность?

##### Решение: 

1. несовместные события независимы $\Leftrightarrow$ хотя бы одно из них $\emptyset$ (недостоверное)

Докажем:

$\Rightarrow$ если события независимы, то $P(A \cap B) = \emptyset \Rightarrow 0 = P(\emptyset) = P(A \cap B) = P(A)\cdot P(B) \Rightarrow P(A) = 0$ или $P(B) = 0 \Rightarrow P(A) = \emptyset$ или $P(B) = \emptyset$

$\Leftarrow$ Пусть $A = \emptyset$, тогда $A \cap B = \emptyset$. Тогда $P(A \cap B) = 0$ и $P(A) \cdot P(B) = 0 \Rightarrow A$ и $B$ - независимые. 

ч.т.д.

2. $P(\prod_{\omega \in \Sigma} \omega) = P(\emptyset) = 0 $ так как $\emptyset \in \Sigma$ 

$\prod_{\omega \in \Sigma} P(\omega) = P(\emptyset) \cdot \prod_{\omega \in \Sigma, \omega \neq \emptyset} P(\omega) = 0 \cdot \prod_{\omega \in \Sigma, \omega \neq \emptyset} P(\omega) = 0$

$\Rightarrow$ независимы


3. В общем случае, могут быть зависимыми и независимыми. Приведем примеры:

зависимые: Пусть A = Выпадение орла, B = Выпадение решки. $P(A) = \frac{1}{2}, P(B) = \frac{1}{2}, P(A \cap B) = P(\emptyset) = 0 \Rightarrow P(A) \cdot P(B) \neq  P(A \cap B)$

независимые: Пусть A = Выпадение орла на 1 монете, B = Выпадение решки на 2 монете. $P(A) = \frac{1}{2}, P(B) = \frac{1}{2}, P(A \cap B) = \frac{1}{4}  \Rightarrow P(A) \cdot P(B) = P(A \cap B)$

#### 1.2 Случайный вектор и числовые характеристики

##### 1. Пусть $$f_{\xi}(x, y) = \frac{e^{-2 | y |}}{\pi (1 + x^2)}$$ Является ли данная функция плотностью распределения случайного вектора? 

##### Решение:

Для того, чтобы функция являлась плотностью некоторого распределения случайного вектора, она должна:

1. быть неотрицательной (так как функция распределения является не убывающей — возрастает, либо остается постоянной, что следует из определения функции распределения $\Rightarrow$ значит её производная всегда будет величиной неотрицательной)

2. значение интеграла по $\mathbb{R^2}$ равно 1


Рассмотрим неотрицательность функции: $$e^{-2 | y |} \geq 0$$ $$\pi (1 + x^2) > 0 $$ 
$$ \Rightarrow f_{\xi}(x,y) \geq 0 \forall x, y \in \mathbb{R}$$ 

Возьмем интеграл: 

$\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \frac{e^{-2 | y |}}{\pi (1 + x^2)} \: \mathrm{d}x \mathrm{d}y = \frac{1}{\pi} \int_{-\infty}^{\infty} e^{-2 | y |} \: \mathrm{d}y \int_{-\infty}^{\infty} \frac{1}{(1 + x^2)} \: \mathrm{d}x =  \frac{1}{\pi} \int_{-\infty}^{\infty} e^{-2 | y |} \: \mathrm{d}y (\arctan x |^{\infty}_{-\infty}) = \frac{1}{\pi} \int_{-\infty}^{\infty} e^{-2 | y |} \: \mathrm{d}y (\frac{\pi}{2} - (- \frac{\pi}{2})) = \int_{-\infty}^{\infty} e^{-2 | y |} \: \mathrm{d}y = |$ так как функция четная $|=$

$= 2 \int_{0}^{\infty} e^{-2y} \: \mathrm{d}y = 2 (-\frac{1}{2} e^{-2y}) |^{\infty}_0 = 0 - (-1) = 1$


Да, функция является плотностью распределения случайного вектора

$E\eta_2 = \int_{-1}^{1} y \cdot \frac{1}{\pi \cdot \sqrt{1 - y^2}} \: \mathrm{d}y = |$ так как функция нечетная $| = 0$

Для ковариационной матрицы рассчитаем дисперсию и коэффициент ковариации:

$D\eta_1 = E\eta_1^2 - (E\eta_1)^2 = E\eta_1^2 = \int_{-1}^{1} y^2 \cdot \frac{1}{\pi \cdot \sqrt{1 - y^2}} \: \mathrm{d}y = |$ так как функция четная $| = \frac{2}{\pi} \int_{0}^{1}  \frac{y^2}{ \cdot \sqrt{1 - y^2}} \: \mathrm{d}y = \begin{vmatrix} 
y = sin\theta \\ 
\sqrt{1 - y^2} = cos\theta \\
\mathrm{d} y = cos\theta \mathrm{d} \theta
\end{vmatrix} = \frac{2}{\pi} \int_{0}^{\frac{\pi}{2}} \frac{sin^2\theta}{cos\theta} \cdot cos\theta  \: \mathrm{d} \theta = \frac{2}{\pi} \int_{0}^{\frac{\pi}{2}} sin^2\theta \: \mathrm{d} \theta = \frac{2}{\pi} \int_{0}^{\frac{\pi}{2}} \frac{1 - cos2\theta}{2} \: \mathrm{d} \theta  = \frac{1}{\pi} \int_{0}^{\frac{\pi}{2}} 1 - cos2\theta \: \mathrm{d} \theta = \frac{1}{\pi} \cdot (\theta - \frac{sin2\theta}{2})^{\frac{\pi}{2}}_{0} = \frac{1}{\pi} \cdot (\frac{\pi}{2} - 0 - 0 + 0) = \frac{1}{2} $  


$D\eta_2 = E\eta_2^2 - (E\eta_2)^2 = E\eta_2^2 = \int_{-1}^{1} y^2 \cdot \frac{1}{\pi \cdot \sqrt{1 - y^2}} \: \mathrm{d}y = \frac{1}{2}$

$cov(\eta_1, \eta_2) = E\eta_1\eta_2 - E\eta_1 \cdot E\eta_2 =  E\eta_1\eta_2 = \int_{-1}^{1} \int_{-1}^{1} \frac{x}{\sqrt{1 - x^2}} \cdot \frac{y}{\sqrt{1 - y^2}} \: \mathrm{d}x \: \mathrm{d} y = 0$

Получаем следующую матрицу ковариации:

$$
\begin{bmatrix}
\frac{1}{2} & 0 \\
\\
0 & \frac{1}{2}
\end{bmatrix}
$$


Коэффициент кореляции $corr(\eta_1, \eta_2) = \frac{cov(\eta_1, \eta_2)}{\sqrt{D\eta_1 \cdot D\eta_2}} = 0$

Матрица корреляции тогда:

$$
\begin{bmatrix}
1 & 0 \\
\\
0 & 1
\end{bmatrix}
$$

##### 5. Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если $\xi ∼ Exp_2$   и   $\eta ∼ U_{0,1}$.

##### Решение:

Запишем функции распределения случайных величин:

$$f_{\xi}(x) = \begin{cases}
2 \cdot e^{-2x}, & \quad x \geq 0 \\ 
0, & \quad x < 0
\end{cases}
$$

$$
f_{\eta}(y)= \begin{cases}
1, & \quad 0 \leq y \leq 1 \\ 
0, & \quad \text{else}
\end{cases}
$$ 

Используем формулу свертки для нахождения плонтости распределения суммы случайных величин:
$$ f_{\xi + \eta} (x) = \int_{-\infty}^{\infty} f_{\xi}(y)f_{\eta}(x-y) \: \mathrm{d} y$$

Ненулевое значение подинтегрального выражения получаем при выполнении следующих требований:

$$
 \begin{cases}
y \geq 0 \\
0 \leq x - y \leq 1 
\end{cases}
\Leftrightarrow
 \begin{cases}
y \geq 0 \\
-1 \leq y - x \leq 0
\end{cases}
\Leftrightarrow
 \begin{cases}
y \geq 0 \\
x - 1 \leq y \leq x 
\end{cases}
$$ 

Определим границы интегрирования, при которых интеграл  $f_{\xi + \eta} (x) = \int_{-\infty}^{\infty} f_{\xi}(y)f_{\eta}(x-y) \: \mathrm{d} y$ не принимает нулевое значение

При $x < 0:$ 

$$\begin{cases}
y \geq 0 \\
x - 1 \leq y \leq x
\end{cases}  - \text{система неравенств не имеет решения}  \Rightarrow f_{\xi + \eta} (x)  = 0$$

При  $0 \leq x \leq 1:$ границы интегрирования   $0 \leq y \leq x$ 

$$f_{\xi + \eta} (x) = \int_{0}^{x} f_{\xi}(y)f_{\eta}(x-y) \: \mathrm{d}y  =  \int_{0}^{x} 2e^{-2y} \cdot 1 \: \mathrm{d}y = (-e^{-2y})|_0^x = -e^{-2x} -(-1) = -e^{-2x} + 1$$

При   $x > 1:$ границы интегрирования  $x - 1 \leq y \leq x$ 

$$f_{\xi + \eta} (x) = \int_{x - 1}^{x} f_{\xi}(y)f_{\eta}(x-y) \: \mathrm{d}y  =  \int_{x - 1}^{x} 2e^{-2y} \cdot 1 \: \mathrm{d}y = (-e^{-2y})|_{x - 1}^x =  -e^{-2x} + e^{-2(x-1)} = e^{-2x}(e^2 - 1)$$



Тогда $$f_{\xi + \eta}(x) = 
\begin{cases}
0, & \quad x < 0 \\
1 - e^{-2x}, & \quad 0 \leq x \leq 1 \\
e^{-2x}(e^2 - 1), & \quad x > 1
\end{cases}$$
